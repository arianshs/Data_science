✅**Scaling for our data**

```python
from sklearn.preprocessing import StandardScaler ,  MinMaxScaler, RobustScaler

df_ss = df.copy()
df_mm = df.copy()
df_rs = df.copy()

ss = StandardScaler()
mm = MinMaxScaler()
rs = RobustScaler()


# behtare aval cat aram ba onehot va ordinal numeric mikardim bad
#kole column haro standard mikardim
#bayad dune dune barresi konim ke koduma one hot shan va ordinal


df_ss[num_col] = ss.fit_transform(df_ss[num_col]) # mean = 0 ,[-∞,+∞]
df_mm[num_col] = mm.fit_transform(df_mm[num_col]) # [0,1]
df_rs[num_col] = rs.fit_transform(df_rs[num_col]) # (x - Q2 )/ IQR

```

✅**Outlier Handling**

in z-score :

  *	حدود 68٪ داده‌ها بین z = -1 و z = +1 هستند (وسط ±1 انحراف معیار)
	*	حدود 95٪ داده‌ها بین z = -2 و z = +2 هستند
	*	حدود 99.7٪ داده‌ها بین z = -3 و z = +3 هستند

--


```python
z_score = np.abs((df[num_col] - df[num_col].mean())/ df[num_col].std())
outliers = (z_score > 2.5).any(axis=1) # mitunim threshold am bezarim baraye barresi features a
# axis 1 miad radif haro barresi mikone .. .any hadeaghal yek feature out bashe True mige
df[outliers]

```

```python
z_score = np.abs((df[num_col] - df[num_col].mean())/ df[num_col].std())
outliers = (z_score > 1.5).all(axis=1) #inja all hame feature haro barresi mikone va agar hame out budan True (100% )
df[outliers].shape
df[~outliers].shape #unai ke outlier nistan
```

```python
#IQR method (raveshe asli)

Q1 = df[num_col].quantile(0.25)
Q3 = df[num_col].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[num_col] < (Q1 - 1.5 * IQR)) | (df[num_col] > (Q3 + 1.5 * IQR))).any(axis=1)
# 1.5 standarde ama baraye taghire sensivity mishe avaz kard

print(f'Number of outliers detected by IQR method: {outliers.sum()}')
print(f'Percentage of outliers: {outliers.sum()/len(df)*100:.2f}%')

df[outliers].shape
```

```python
from scipy.stats import skew , kurtosis, jarque_bera, gaussian_kde
skew(df[num_col])

# --- Metric helper ---
def metrics(vec):
  sk= skew(vec, nan_policy='omit')
  ku = kurtosis(vec , fisher=True, nan_policy='omit') # 0 = normal
  jb_stat , jb_p = jarque_bera(vec)
  return sk, ku, jb_p
```

```python

# داده نمونه
data = df['TotalCharges'].dropna()

# ایجاد Figure
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")  # پس‌زمینه و grid مرتب

# رسم هیستوگرام با تعداد واقعی
sns.histplot(data, bins=30, stat="count", color='skyblue', alpha=0.6, edgecolor='black')

# محاسبه KDE
kde = gaussian_kde(data)
x = np.linspace(min(data), max(data), 1000)
kde_scaled = kde(x) * len(data) * (max(data)-min(data))/30  # مقیاس برای هماهنگی با histogram

# رسم خط منحنی حرفه‌ای
plt.plot(x, kde_scaled, color='red', lw=2, label='KDE Curve')
plt.fill_between(x, 0, kde_scaled, color='red', alpha=0.3)

# تنظیمات عنوان و محور
plt.title('Distribution of TotalCharges (Bell Shape with Count)', fontsize=16)
plt.xlabel('TotalCharges', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend()
plt.show()
```
